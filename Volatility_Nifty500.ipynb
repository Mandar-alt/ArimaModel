```python
import pandas as pd
import numpy as np
```

```python
nifty500_stocks = pd.read_csv('Nifty_500.csv')
```

```python
print(nifty500_stocks[['Date', 'Adj Close']])
```

```python
# Calculated simple return of the adjusted close column
nifty500_stocks['simple_return'] = nifty500_stocks['Adj Close'].pct_change()
print(nifty500_stocks[['Date', 'Adj Close', 'simple_return']])
```

Calculating Stock Volatility

```python

# Calculating Stock Volatility
nifty500_stocks['daily_std_dev'] = nifty500_stocks['simple_return'].std()
print(nifty500_stocks[['Date', 'Adj Close', 'daily_std_dev']])
```

```python
# Calculate annualized standard deviation
nifty500_stocks['annualized_std_dev'] = nifty500_stocks['daily_std_dev'] * np.sqrt(252)
print(nifty500_stocks[['Date', 'Adj Close', 'annualized_std_dev']])
```

```python
print("Annunal Stock volatilty for 0 index : {0}".format(nifty500_stocks['annualized_std_dev'][0]))
```

```python

# Set threshold for high and low volatility
high_volatility_threshold = 0.3
low_volatility_threshold = 0.1

# Classify stocks based on volatility
high_volatility_stocks = nifty500_stocks[nifty500_stocks['annualized_std_dev'] > high_volatility_threshold]
low_volatility_stocks = nifty500_stocks[nifty500_stocks['annualized_std_dev'] < low_volatility_threshold]

print("High Volatility Stocks: ")
print(high_volatility_stocks)

print("\nLow Volatility Stocks:")
print(low_volatility_stocks)
```

```python
import pandas as pd
import numpy as np
import yfinance as yf

nifty_500 = ['ZOMATO.NS', 'AWL.NS', 'J&KBANK.NS', 'PAYTM.NS', 'PCBL.NS',
                     'JUBLINGREA.NS', 'NAUKRI.NS', 'BIKAJI.NS', 'SUZLON.NS',
                     'CLEAN.NS', 'NYKAA.NS', 'RKFORGE.NS', 'TTML.NS',
                     'AMBER.NS', 'CAPLIPOINT.NS', 'HSCL.NS', 'DEEPAKFERT.NS',
                     'POLICYBZR.NS', 'IEX.NS', 'SAFARI.NS', 'SONACOMS.NS',
                     'ZFCVINDIA.NS', 'GLAXO.NS', 'APLLTD.NS', 'JUBLFOOD.NS',
                     'AJANTPHARM.NS', 'DELHIVERY.NS', 'INDIACEM.NS',
                     'FIVESTAR.NS', 'DIVISLAB.NS', 'STLTECH.NS', 'CASTROLIND.NS',
                     'INOXWIND.NS', 'IDEA.NS', 'PEL.NS', 'JWL.NS', 'GILLETTE.NS',
                     'ASTERDM.NS', 'IDBI.NS', 'JBCHEPHARM.NS', 'CEATLTD.NS',
                     'ANGELONE.NS', 'FDC.NS', 'AFFLE.NS', 'CENTURYPLY.NS',
                     'DEVYANI.NS', 'EQUITASBNK.NS', 'PAGEIND.NS', 'PGHH.NS',
                     'WELCORP.NS', 'MGL.NS', 'ALKYLAMINE.NS', 'MINDACORP.NS',
                     'CHAMBLFERT.NS', 'HDFCBANK.NS', 'BIOCON.NS', 'NBCC.NS',
                     'TORNTPHARM.NS', 'RAMCOCEM.NS', 'AETHER.NS', 'JKPAPER.NS',
                     'LTF.NS', 'HAPPYFORGE.NS', 'DRREDDY.NS', 'UTIAMC.NS',
                     'OIL.NS', 'AUROPHARMA.NS', 'ZEEL.NS', 'PPLPHARMA.NS',
                     'POONAWALLA.NS', 'SUVENPHAR.NS', 'CELLO.NS', 'MUTHOOTFIN.NS',
                     'RAILTEL.NS', 'VARROC.NS', 'PIDILITIND.NS', 'ALKEM.NS',
                     'GLENMARK.NS', 'SUNPHARMA.NS', 'MOTILALOFS.NS',
                     'SCHNEIDER.NS', 'ZYDUSLIFE.NS', 'SHRIRAMFIN.NS',
                     'ARE&M.NS', 'KOTAKBANK.NS', 'BAJAJHLDNG.NS', 'EPL.NS',
                     'ABFRL.NS', 'AAVAS.NS', 'DEEPAKNTR.NS', 'LUPIN.NS',
                     'LLOYDSME.NS', 'UNITDSPR.NS', 'RBA.NS', 'ASIANPAINT.NS',
                     'TORNTPOWER.NS', 'HINDPETRO.NS', 'MFSL.NS', 'ALLCARGO.NS',
                     'CDSL.NS', 'DOMS.NS', 'NESTLEIND.NS', 'CANFINHOME.NS',
                     'GLAND.NS', 'TANLA.NS', 'ASHOKLEY.NS', 'ABBOTINDIA.NS',
                     'MHRIL.NS', 'PRSMJOHNSN.NS', 'POLYMED.NS', 'KPITTECH.NS',
                     'BALRAMCHIN.NS', 'PNBHOUSING.NS', 'COROMANDEL.NS',
                     'ASTRAZEN.NS', 'JINDALSAW.NS', 'METROPOLIS.NS', 'RITES.NS',
                     'GSPL.NS', 'APOLLOHOSP.NS', 'LEMONTREE.NS', 'NATCOPHARM.NS',
                     'VGUARD.NS', 'SUNTV.NS', '3MINDIA.NS', 'RAINBOW.NS',
                     'COLPAL.NS', 'CIPLA.NS', 'DATAPATTNS.NS', 'TRITURBINE.NS',
                     'LALPATHLAB.NS', 'HUDCO.NS', 'BANDHANBNK.NS', 'KPRMILL.NS',
                     'JMFINANCIL.NS', 'MASTEK.NS', 'BRITANNIA.NS', 'RENUKA.NS',
                     'UBL.NS', 'LATENTVIEW.NS', 'CYIENT.NS', 'ECLERX.NS',
                     'APTUS.NS', 'GMMPFAUDLR.NS', 'GUJGASLTD.NS', 'DIXON.NS',
                     'FORTIS.NS', 'KRBL.NS', 'CHEMPLASTS.NS', 'GICRE.NS',
                     'LXCHEM.NS', 'JUSTDIAL.NS', 'BANKINDIA.NS', 'NETWORK18.NS',
                     'AARTIIND.NS', 'SPARC.NS', 'BERGEPAINT.NS', 'RHIM.NS',
                     'SUNDRMFAST.NS', 'NCC.NS', 'GAEL.NS', 'VBL.NS',
                     'TATACOMM.NS', 'UPL.NS', 'CHOLAFIN.NS', 'POLYCAB.NS',
                     'BATAINDIA.NS', 'IGL.NS', 'ADANIPORTS.NS', 'CSBBANK.NS',
                     'UCOBANK.NS', 'INDIANB.NS', 'TATAPOWER.NS', 'PRAJIND.NS',
                     'AIAENG.NS', 'MAZDOCK.NS', 'WHIRL.NS', 'TATAELXSI.NS',
                     'BAYERCROP.NS', 'M&MFIN.NS', 'LTTS.NS', 'GPIL.NS',
                     'HAVELLS.NS', 'TMB.NS', 'CHALET.NS', 'JSL.NS',
                     'JYOTHYLAB.NS', 'DCMSHRIRAM.NS', 'ADANIENSOL.NS',
                     'MSUMI.NS', 'FINCABLES.NS', 'AUBANK.NS', 'SANOFI.NS',
                     'CRISIL.NS', 'SUMICHEM.NS', 'BALAMINES.NS', 'HONAUT.NS',
                     'ATUL.NS', 'SIGNATURE.NS', 'HOMEFIRST.NS', 'BPCL.NS',
                     'ANANDRATHI.NS', 'REDINGTON.NS', 'HFCL.NS', 'INDHOTEL.NS',
                     'MANKIND.NS', 'KIMS.NS', 'TATATECH.NS', 'BORORENEW.NS',
                     'GODFRYPHLP.NS', 'GODREJCP.NS', 'EXIDEIND.NS',
                     'EASEMYTRIP.NS', 'ERIS.NS', 'IRFC.NS', 'JIOFIN.NS',
                     'ITI.NS', 'DALBHARAT.NS', 'ICICIPRULI.NS', 'MAHABANK.NS',
                     'KAJARIACER.NS', 'TATAMOTORS.NS', 'HGS.NS', 'RELAXO.NS',
                     'PERSISTENT.NS', 'FINEORG.NS', 'GSFC.NS', 'EICHERMOT.NS',
                     'TRIVENI.NS', 'GREENPANEL.NS', 'CAMS.NS', 'SAPPHIRE.NS',
                     'SHARDACROP.NS', 'APEX.NS', 'BRIGADE.NS', 'UJJIVANSFB.NS',
                     'PNCINFRA.NS', 'KARURVYSYA.NS', 'MIDHANI.NS', 'AHL.NS',
                     'INDIAMART.NS', 'CROMPTON.NS', 'RVNL.NS', 'SUPREMEIND.NS',
                     'HGS.NS']

results = []

# Loop through each stock to calculate volatility, variance, fetch enterprise value, operating margin, and earnings per share
for ticker in nifty_500:
    try:
        # Fetch stock info
        stock = yf.Ticker(ticker)
        info = stock.info

        # Debugging: Print the fetched info for each stock
        print(f"Info for {ticker}: {info}")

        # Calculate volatility from daily returns
        data = stock.history(period="max")  # Get max available data
        returns = data['Close'].pct_change().dropna()  # Calculate daily returns

        # Calculate volatility (annualized)
        volatility = returns.std() * np.sqrt(252)

        # Calculate variance (daily)
        variance = returns.var()

        # Get enterprise value components
        market_cap = info.get('marketCap')
        total_debt = info.get('totalDebt')
        cash_and_equivalents = info.get('totalCash')

        # Calculate enterprise value
        if market_cap is not None and total_debt is not None and cash_and_equivalents is not None:
            enterprise_value = market_cap + total_debt - cash_and_equivalents
        else:
            enterprise_value = None

        # Fetch operating margin
        operating_margin = info.get('operatingMargins')

        # Fetch earnings per share (EPS)
        earnings_per_share = info.get('epsTrailingTwelveMonths')

        # Calculate Enterprise Value / Operating Margin ratio
        if enterprise_value is not None and operating_margin not in [None, 0]:
            ev_to_operating_margin = enterprise_value / operating_margin
        else:
            ev_to_operating_margin = None

        # Append the result to the DataFrame
        results.append({
            'Ticker': ticker,
            'Enterprise Value': enterprise_value,
            'Market Cap': market_cap,
            'Volatility': volatility,
            'Variance': variance,
            'Operating Margin': operating_margin,
            'Earnings Per Share': earnings_per_share,
            'EV/Operating Margin': ev_to_operating_margin
        })

    except Exception as e:
        print(f"Couldn't fetch data for {ticker}: {e}")

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Print or save the results
print(results_df.head())

# Save the results to CSV if needed
results_df.to_csv('Nifty_500_with_yfinfo.csv', index=False)
```
